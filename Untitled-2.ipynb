## The code scrapes news headlines from the Economic Times archive for a given date using the archive URL format.

```python


# Function to scrape news headlines, links, and date dynamically
def get_news_data(year, month, starttime):
    url = f"https://economictimes.indiatimes.com/archivelist/year-{year},month-{month},starttime-{starttime}.cms"
    
    # Send a request to the website
    response = requests.get(url, headers={"User-Agent": "Mozilla/5.0"})
    
    # Check if the request was successful
    if response.status_code != 200:
        print(f"Failed to retrieve data: {response.status_code}")
        return pd.DataFrame()  # Return empty DataFrame if failed
    
    # Parse the HTML content
    soup = BeautifulSoup(response.text, "html.parser")
    
    # Extract the date from the <td> tag
    date_tag = soup.find("td", class_="contentbox5")
    if date_tag:
        date_text = date_tag.get_text(strip=True)
        date = date_text.split(">")[-1].strip()  # Extract last part (e.g., "01 Feb, 2023")
    else:
        date = "Unknown"

    # Find the section containing headlines
    content_section = soup.find("ul", class_="content")
    
    if not content_section:
        print("Could not find news headlines section.-",year,month,starttime)
        return pd.DataFrame()  # Return empty DataFrame if no content section found

    # Extract news headlines and links
    news_data = []
    for li in content_section.find_all("li"):
        a_tag = li.find("a")
        if a_tag:
            headline = a_tag.get_text(strip=True)
            link =a_tag["href"]  # Ensure full URL
            news_data.append([date, headline, link])

    # Convert to DataFrame
    df = pd.DataFrame(news_data, columns=["Date", "Headline", "Link"])
    
    return df

# Example: Scrape news headlines for 01-Feb-2023
year, month, starttime = 2023, 2, 44958  # Starttime for 01-Feb-2023

df_news = get_news_data(year, month, starttime)

# Display the DataFrame
print(df_news)

# Save to CSV (optional)
#df_news.to_csv("news_feb_01_2023.csv", index=False)
```

## This script automates scraping daily news headlines from Times of India from Jan 1, 2013 to Jan 31, 2025 using multithreading for faster performance.

```python


start_date = date(2013, 1, 1)
end_date = date(2025, 1, 31)
delta = timedelta(days=1)

starttime = 41275
max_no_data_limit = 5

news_list = []
no_data_count = 0  

# Function to fetch data for a specific date
def fetch_news(year, month, day, starttime):
    df_day = get_news_data(year=year, month=month, day=day, starttime=starttime)
    print(f'extracting news-',year,month,day,starttime)
    if df_day is None or df_day.empty:
        print(f"No data for {year}-{month}-{day}. Skipping.")
        return None  # Return None if no data
    return df_day  # Return the dataframe if data exists

# Collect all dates first
date_list = [(start_date + timedelta(days=i), starttime + i) for i in range((end_date - start_date).days + 1)]

# Use ThreadPoolExecutor to run requests in parallel
with ThreadPoolExecutor(max_workers=10) as executor:  # Adjust workers based on CPU
    results = executor.map(lambda d: fetch_news(d[0].year, d[0].month, d[0].day, d[1]), date_list)

# Filter out None values and store valid data
news_list = [df for df in results if df is not None]

print("Data collection completed.")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from concurrent.futures import ThreadPoolExecutor, as_completed

# Function to extract "Last Updated Date & Time" from each news link
def get_last_updated_time(url):
    try:
        response = requests.get(url, headers={"User-Agent": "Mozilla/5.0"}, timeout=5)
        
        if response.status_code != 200:
            print(f"Failed to fetch {url} - Status Code: {response.status_code}")
            return None
        
        soup = BeautifulSoup(response.text, "html.parser")
        byline_div = soup.find("div", class_="xf8Pm byline")
        
        if byline_div:
            span_tag = byline_div.find("span")
            if span_tag:
                return span_tag.get_text(strip=True)  # Example: "Updated: Jan 24, 2017, 09:30 IST"
        
        return None  # Return None if timestamp not found
        
    except Exception as e:
        print(f"Error fetching {url}: {e}")
        return None

# Function to scrape multiple URLs concurrently
def scrape_news_links(links, max_threads=10):
    last_updated_times = []
    
    with ThreadPoolExecutor(max_threads) as executor:
        future_to_url = {executor.submit(get_last_updated_time, url): url for url in links}
        
        for future in as_completed(future_to_url):
            url = future_to_url[future]
            try:
                last_updated_times.append(future.result())
            except Exception as e:
                print(f"Exception for {url}: {e}")
                last_updated_times.append(None)
    
    return last_updated_times

# Example DataFrame with multiple news article links


# Extract "Last Updated" date & time using multithreading
df["Last_Updated"] = scrape_news_links(df["Link"].tolist(), max_threads=10)

# Display the updated DataFrame
print(df)
```

## To insert news data from a Pandas DataFrame into a MySQL database table called News.

```python
import mysql.connector
from mysql.connector import Error

def insert_into_mysql(df):
    try:
        # Connect to MySQL database
        connection = mysql.connector.connect(
            host="localhost",
            user="root",
            password="watambe13aw",  # Change this
            database="news"
        )

        if connection.is_connected():
            cursor = connection.cursor()
            
            # SQL query for inserting data
            sql = """INSERT INTO News (Date, Headline, Link, Last_Updated)
                     VALUES (%s, %s, %s, NOW())"""  # `NOW()` for auto timestamp

            # Convert Date column to string (if needed)
            data = [
                (row["Date"].strftime('%Y-%m-%d') if hasattr(row["Date"], "strftime") else row["Date"],
                 row["Headline"], row["Link"])
                for _, row in df.iterrows()
            ]

            # Bulk insert
            cursor.executemany(sql, data)

            # Commit changes
            connection.commit()
            print(f"{cursor.rowcount} rows inserted successfully!")

    except Error as e:
        print(f"Error: {e}")

    finally:
        if connection.is_connected():
            cursor.close()
            connection.close()
            print("MySQL connection closed.")

# Call the function
insert_into_mysql(df)
```
